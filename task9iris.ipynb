Pair plots


A pairs plot allows us to see both distribution of single variables and relationships between two variables.

For example, let’s say we have four features ‘sepal length’, ‘sepal width’, ‘petal length’ and ‘petal width’ in our iris dataset. In that case, we will have 4C2 plots i.e. 6 unique plots.

The pairs in this case will be :

sepal length, sepal width
sepal length, petal length
sepal length, petal width
sepal width, petal length
sepal width, petal width
petal length, petal width
So, here instead of trying to visualize four dimensions which is not possible. We will look into 6 2D plots and try to understand the 4-dimensional data in the form of a matrix.

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]


sns.set_style("whitegrid")
sns.pairplot(iris_df, hue="species", height=2)
plt.show()
```

Limitation of Pair plot:-

If you have d features, you will have a pair plot of size dxd cells where each cell is a plot between a pair of features.So, pair plots are hard to use when we have high dimensional data. For high dimensional data we can use PCA, t-SNE.

A histogram is an accurate graphical representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable (quantitative variable).To construct a histogram, the first step is to bin the range of values — that is, divide the entire range of values into a series of intervals — and then count how many values fall into each interval.
```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

iris_setosa = iris_df[iris_df["species"] == "setosa"]
iris_virginica = iris_df[iris_df["species"] == "virginica"]
iris_versicolor = iris_df[iris_df["species"] == "versicolor"]


counts, bin_edges = np.histogram(iris_setosa['petal length (cm)'], bins=10, density=True)

print("Counts:", counts)
print("Bin edges:", bin_edges)
```

```python
print("Prrinting the counts for histogram\n",counts)
pdf = counts/(sum(counts))
print("Sum of count is\n",sum(counts))
print("Printing Edges of bins \n",bin_edges)
print("PDF is as followes  \n",pdf);

plt.gca().legend(('Pdf'))
plt.title('PDF and PDF For iris_setosa')
plt.xlabel("Petal length")
plt.ylabel("Percentage")
plt.plot(bin_edges[1:],pdf)
```

Calculating CDF Points

```python
cdf = np.cumsum(pdf)
print("CDF Is as followes\n",cdf)
plt.gca().legend(('Cdf'))
plt.title('CDF For iris_setosa')
plt.xlabel("Petal length")
plt.ylabel("Percentage")

plt.plot(bin_edges[1:],cdf)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]


iris_setosa = iris_df[iris_df["species"] == "setosa"]


counts, bin_edges = np.histogram(iris_setosa['petal length (cm)'], bins=10, density=True)


pdf = counts / np.sum(counts)
print("PDF:", pdf)


cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf, label='PDF')
plt.plot(bin_edges[1:], cdf, label='CDF')
plt.legend()
plt.title('PDF and CDF for Iris Setosa')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Percentage')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]


iris_setosa = iris_df[iris_df["species"] == "setosa"]


counts, bin_edges = np.histogram(iris_setosa['petal length (cm)'], bins=10, density=True)
pdf = counts
cdf = np.cumsum(pdf) * np.diff(bin_edges)  # Adjust CDF calculation
plt.plot(bin_edges[1:], pdf, label='PDF (10 bins)')
plt.plot(bin_edges[1:], cdf, label='CDF (10 bins)')


counts, bin_edges = np.histogram(iris_setosa['petal length (cm)'], bins=20, density=True)
pdf = counts
cdf = np.cumsum(pdf) * np.diff(bin_edges)  # Adjust CDF calculation
plt.plot(bin_edges[1:], pdf, label='PDF (20 bins)')

plt.legend()
plt.title('PDF and CDF for Iris Setosa')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Percentage')
plt.show()
```

PDF and CDF For all species

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

iris_setosa = iris_df[iris_df["species"] == "setosa"]
iris_virginica = iris_df[iris_df["species"] == "virginica"]
iris_versicolor = iris_df[iris_df["species"] == "versicolor"]

def plot_pdf_cdf(data, species_name, color):
    counts, bin_edges = np.histogram(data['petal length (cm)'], bins=10, density=True)
    pdf = counts
    cdf = np.cumsum(pdf) * np.diff(bin_edges)  # Adjust for bin width
    plt.plot(bin_edges[1:], pdf, color=color, label=f'PDF {species_name}')
    plt.plot(bin_edges[1:], cdf, color=color, linestyle='--', label=f'CDF {species_name}')


plot_pdf_cdf(iris_setosa, 'Setosa', 'blue')
plot_pdf_cdf(iris_virginica, 'Virginica', 'green')
plot_pdf_cdf(iris_versicolor, 'Versicolor', 'red')

plt.title('PDF and CDF for Iris Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density / Percentage')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]


iris_setosa = iris_df[iris_df["species"] == "setosa"]
iris_virginica = iris_df[iris_df["species"] == "virginica"]
iris_versicolor = iris_df[iris_df["species"] == "versicolor"]

print("Means:")
print("Setosa Mean:", np.mean(iris_setosa["petal length (cm)"]))
print("Setosa Mean with outlier:", np.mean(np.append(iris_setosa["petal length (cm)"], 50)))
print("Virginica Mean:", np.mean(iris_virginica["petal length (cm)"]))
print("Versicolor Mean:", np.mean(iris_versicolor["petal length (cm)"]))

print("\nVariances:")
print("Setosa Variance:", np.var(iris_setosa["petal length (cm)"], ddof=0))
print("Virginica Variance:", np.var(iris_virginica["petal length (cm)"], ddof=0))
print("Versicolor Variance:", np.var(iris_versicolor["petal length (cm)"], ddof=0))

print("\nStandard Deviations:")
print("Setosa Std Dev:", np.std(iris_setosa["petal length (cm)"], ddof=0))
print("Virginica Std Dev:", np.std(iris_virginica["petal length (cm)"], ddof=0))
print("Versicolor Std Dev:", np.std(iris_versicolor["petal length (cm)"], ddof=0))

```

```python

import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from statsmodels import robust

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

# Filter by species
iris_setosa = iris_df[iris_df["species"] == "setosa"]
iris_virginica = iris_df[iris_df["species"] == "virginica"]
iris_versicolor = iris_df[iris_df["species"] == "versicolor"]

# Median calculations
print("\nMedians:")
print("Setosa Median:", np.median(iris_setosa["petal length (cm)"]))
print("Setosa Median with outlier:", np.median(np.append(iris_setosa["petal length (cm)"], 50)))
print("Virginica Median:", np.median(iris_virginica["petal length (cm)"]))
print("Versicolor Median:", np.median(iris_versicolor["petal length (cm)"]))

# Quantiles (25th, 50th, 75th)
print("\nQuantiles:")
print("Setosa Quantiles:", np.percentile(iris_setosa["petal length (cm)"], np.arange(0, 100, 25)))
print("Virginica Quantiles:", np.percentile(iris_virginica["petal length (cm)"], np.arange(0, 100, 25)))
print("Versicolor Quantiles:", np.percentile(iris_versicolor["petal length (cm)"], np.arange(0, 100, 25)))

# 90th Percentiles
print("\n90th Percentiles:")
print("Setosa 90th Percentile:", np.percentile(iris_setosa["petal length (cm)"], 90))
print("Virginica 90th Percentile:", np.percentile(iris_virginica["petal length (cm)"], 90))
print("Versicolor 90th Percentile:", np.percentile(iris_versicolor["petal length (cm)"], 90))

# Median Absolute Deviation
print("\nMedian Absolute Deviation:")
print("Setosa MAD:", robust.mad(iris_setosa["petal length (cm)"]))
print("Virginica MAD:", robust.mad(iris_virginica["petal length (cm)"]))
print("Versicolor MAD:", robust.mad(iris_versicolor["petal length (cm)"]))
```

Box-plot and whisker

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary. It does not show a distribution in as much detail as a stem and leaf plot or histogram does, but is especially useful for indicating whether a distribution is skewed and whether there are potential unusual observations (outliers) in the data set.

Box-plot with whiskers: another method of visualising the 1-D scatter plot more intuitive

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris


iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]

sns.boxplot(x="species", y="petal length (cm)", data=iris_df)
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```
