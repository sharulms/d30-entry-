1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")  # Adjust the file path accordingly

# Check the first few rows to understand the structure
print(df.head())

# Assume the column for teams is named 'Team'
# Calculate the distribution of employees across each team
team_distribution = df['Team'].value_counts()

# Calculate the total number of employees
total_employees = len(df)

# Calculate the percentage split relative to the total number of employees
percentage_split = (team_distribution / total_employees) * 100

# Create a DataFrame to display results
distribution_df = pd.DataFrame({
    'Number of Employees': team_distribution,
    'Percentage of Total Employees': percentage_split
})

# Display the results
print(distribution_df)
```

 2 Segregate employees based on their positions within the company.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")  # Adjust the file path accordingly

# Check the first few rows to understand the structure
print(df.head())

# Assume the column for positions is named 'Position'
# Group by position and count the number of employees in each position
position_distribution = df['Position'].value_counts()

# Create a DataFrame to display results
position_df = pd.DataFrame({
    'Number of Employees': position_distribution
}).reset_index()

# Rename the columns for clarity
position_df.columns = ['Position', 'Number of Employees']

# Display the results
print(position_df)
```

3. Identify the predominant age group among employees.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")  # Adjust the file path accordingly

# Check the first few rows to understand the structure
print(df.head())

# Assume the column for ages is named 'Age'
# Define the bins for age groups
bins = [0, 20, 30, 40, 50, 60, 100]  # Age groups: 0-20, 21-30, 31-40, 41-50, 51-60, 61+
labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61+']

# Create a new column 'Age Group' in the DataFrame
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_distribution = df['Age Group'].value_counts()

# Identify the predominant age group
predominant_age_group = age_group_distribution.idxmax()
predominant_count = age_group_distribution.max()

# Create a DataFrame to display results
age_group_df = pd.DataFrame({
    'Number of Employees': age_group_distribution
}).reset_index()

# Rename the columns for clarity
age_group_df.columns = ['Age Group', 'Number of Employees']

# Display the results
print(age_group_df)

print(f"The predominant age group is {predominant_age_group} with {predominant_count}employees.")
```

4. Discover which team and position have the highest salary expenditure.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")  # Adjust the file path accordingly

# Check the first few rows to understand the structure
print(df.head())

# Assume the columns are named 'Team', 'Position', and 'Salary'
# Group by Team and Position, then sum the salaries
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Rename the columns for clarity
salary_expenditure.columns = ['Team', 'Position', 'Total Salary Expenditure']

# Identify the team and position with the highest salary expenditure
highest_expenditure = salary_expenditure.loc[salary_expenditure['Total Salary Expenditure'].idxmax()]

# Display the results
print(salary_expenditure)
print(f"The team '{highest_expenditure['Team']}' and position '{highest_expenditure['Position']}' have the highest salary expenditure of {highest_expenditure['Total Salary Expenditure']}.")
```

5. Investigate if there's any correlation between age and salary, and represent it visually

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")  # Adjust the file path accordingly

# Check the first few rows to understand the structure
print(df.head())

# Assume the columns are named 'Age' and 'Salary'
# Calculate the correlation between Age and Salary
correlation = df['Age'].corr(df['Salary'])
print(f"The correlation between age and salary is: {correlation}")

# Visualize the relationship using a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

1. Employee Distribution Across Teams

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("myexcel.csv.csv")

# Employee distribution
team_distribution = df['Team'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
team_distribution.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution Across Teams')
plt.xlabel('Teams')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

2. Salary Distribution of Employees

```python
import seaborn as sns

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='Position', y='Salary', data=df)
plt.title('Salary Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

3. Predominant Age Group

```python
# Count employees in each age group
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

df = pd.read_csv("myexcel.csv.csv")
age_group_distribution = df['Age'].value_counts()

# Plotting
plt.figure(figsize=(8, 8))
age_group_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Employee Age Group Distribution')
plt.ylabel('')  # Hide y-label
plt.show()

```

4. Team and Position with Highest Salary Expenditure


```python
# Group by Team and Position to get total salary expenditure
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().unstack()

# Plotting
plt.figure(figsize=(12, 8))
sns.heatmap(salary_expenditure, annot=True, fmt='.0f', cmap='YlGnBu', cbar=True)
plt.title('Total Salary Expenditure by Team and Position')
plt.xlabel('Position')
plt.ylabel('Team')
plt.show()

```

5. Correlation Between Age and Salary

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=df, scatter_kws={'alpha':0.5}, line_kws={"color": "red"})
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
